# SECTION 12 — Feature Engineering

### Create Risk Category

```python
df["Risk_Category"] = (
    df["Risk_Score"]
      .apply(
          lambda x:
          "Critical"
          if x >= 80
          else "High"
          if x >= 60
          else "Medium"
          if x >= 40
          else "Low"
      )
)
df[["Risk_Score", "Risk_Category"]].head()
```

### Amount Category

```python
df["Amount_Category"] = pd.cut(
    df["Transaction_Amount"],
    bins=[0,50000,100000,150000,250000],
    labels=["Low","Medium","High","Very High"]
)

df["Amount_Category"].value_counts()
```

### KYC Risk Flag

```python
df["KYC_Risk"] = np.where(
    df["KYC_Status"]=="Incomplete",
    1,
    0
)

df["KYC_Risk"].value_counts()
```

# SECTION 14 — Random Forest Model

```python
# Customer Risk Encoding

df["Customer_Risk_enc"] = (
    df["Customer_Risk"]
      .map({
          "Low":0,
          "Medium":1,
          "High":2
      })
)

# Country Risk Encoding

df["Country_Risk_enc"] = (
    df["Country_Risk"]
      .map({
          "Low":0,
          "Medium":1,
          "High":2
      })
)

# KYC Status Encoding

df["KYC_Status_enc"] = df["KYC_Status"].map({
    "Complete": 0,
    "Pending": 1,
    "Incomplete": 2
})

# Check Results

df[
[
"Customer_Risk",
"Customer_Risk_enc",
"Country_Risk",
"Country_Risk_enc",
"KYC_Status",
"KYC_Status_enc"
]
].head()
```

```python
# Validation Check
df[
[
"Country_Risk",
"Country_Risk_enc"
]
].drop_duplicates()
```

### Features

```python
features = [
    "Transaction_Amount",
    "Risk_Score",
    "Customer_Risk_enc",
    "Country_Risk_enc",
    "KYC_Status_enc",
    "Account_Age_Months"
]

X = df[features]

y = df["SAR_Flag"]
X.head()
```

### Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

### Train Model

```python
model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)
print("Model Trained Successfully")
```

### Predictions

```python
y_pred = model.predict(
    X_test
)
y_pred[:10]
```

### Model Accuracy

```python
from sklearn.metrics import accuracy_score

acc = round(
    accuracy_score(y_test, y_pred) * 100,
    2
)

print("=" * 40)
print(f"Model Accuracy : {acc}%")
print("=" * 40)
```

### Overfitting Check

```python
print("Train Accuracy:",
      round(model.score(X_train, y_train)*100,2))

print("Test Accuracy:",
      round(model.score(X_test, y_test)*100,2))
```

# Section 16A — ROC-AUC Score

```python
y_prob = model.predict_proba(X_test)[:,1]

auc = roc_auc_score(
    y_test,
    y_prob
)

print("ROC AUC Score:", auc)
```

# Section 16B — ROC Curve

```python
fpr, tpr, thresholds = roc_curve(
    y_test,
    y_prob
)

plt.figure(figsize=(8,5))

plt.plot(
    fpr,
    tpr,
    label=f"AUC = {auc:.2f}"
)

plt.plot(
    [0,1],
    [0,1],
    linestyle="--"
)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")

plt.legend()

plt.show()
```

# SECTION 18 — Isolation Forest

### Anomaly Detection

```python
iso = IsolationForest(contamination=0.05, random_state=42)
df["Anomaly"] = iso.fit_predict(
    df[
        [
            "Transaction_Amount",
            "Risk_Score",
            "Customer_Risk_enc",
            "Country_Risk_enc",
            "KYC_Status_enc",
            "Account_Age_Months"
        ]
    ]
)
```

### Convert Labels

```python
df["Anomaly"] = (
    df["Anomaly"]
      .map(
          {
              1: "Normal",
              -1: "Anomaly"
          }
      )
)
```

### Anomaly Counts

```python
df["Anomaly"].value_counts()
```

### Anamoly Profiling

```python
anomaly_profile = (
    df[df["Anomaly"]=="Anomaly"][
        ["Transaction_Amount","Risk_Score"]
    ]
    .describe()
)

anomaly_profile
```

Purpose:
• Isolation Forest was used to identify unusual transaction patterns.

Why Used:
• It detects anomalies without requiring labeled fraud data.

Key Observation:
• A small subset of transactions was classified as anomalous and generally showed higher transaction amounts and risk scores.

Business Impact:
• These transactions can be prioritized for AML investigation and enhanced monitoring.
